## Creating your first network

```python
import networkx as nx

G = nx.Graph()

G.add_node('A')
G.add_nodes_from(['B', 'C'])

G.add_edge('A', 'B')
G.add_edges_from([('B', 'C'), ('A', 'C')])

nx.draw_networkx(G)
```

### Add several nodes and edges at the same time

```python
G.add_edges_from([('B', 'D'), ('C', 'E')])
nx.draw_networkx(G)
```
